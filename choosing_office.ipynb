```python
from haversine import haversine
import pandas as pd
```

```python
haversine((45.8,4.84), (48.86,2.35), unit="m")
```

```python
def scoring_distance(coords1, coords2):
    coords1=tuple(coords1)
    coords2=tuple(coords2)
    
    return 1/(haversine(coords1, coords2))

   
    
    
```

```python
scoring_distance((45.8,4.84), (45.86,4.35))
```

```python
df_offices = pd.read_csv("offices.csv")
```

```python
df_preferences = pd.read_csv ("places.csv")
```

```python
df_offices.head()
```

```python
df_preferences.head()
```

```python
df_offices['total_score'] = 0
for i in range(df_offices.shape[0]):
    for j in range(df_preferences.shape[0]):
            coord1 = (df_offices.loc[i,'lat'], df_offices.loc[i,'lng'])
            coord2 = (df_preferences.loc[j,'lat'], df_preferences.loc[j,'lng'])
            df_offices.loc[i, 'total_score'] += scoring_distance(coord1,coord2)
            
df_offices
        
    
```

```python
df_offices.sort_values(by="total_score", ascending = False)
```
